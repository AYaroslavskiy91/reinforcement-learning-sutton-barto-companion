```python
#| label: fig-polar
#| fig-cap: "A line plot on a polar axis"

import numpy as np
import matplotlib.pyplot as plt

r = np.arange(0,2,.01)
theta = 2*np.pi*r
fix, ax = plt.subplots(
    subplot_kw = {'projection': 'polar'}
)
ax.set_rticks([.5,1,1.5,2])
ax.grid(True)
ax.plot(theta, r)
plt.show()
```
